# filter()

- It filters the sequence(list, tuple,...) with the help of a function which test each of the elements in the sequence is present or not.

```python
# Filter
# syntax: filter(function, sequence/list/tuple..)
letters = ['a', 'b', 'c', 'd' ]

def function_0(variable):
    if variable in letters:
        return True
    else:
        return False

check_seq = ['a', 'c', 'v', 'z', 'x']
filtered = filter(function_0, check_seq)

for i in filtered:
    print(i)
```

```python
# Scenario : Want a filtered list which gives me even numbers
seq = [0, 1, 2, 3, 4, 5, 6]
result = lambda x: x % 2 == 0

filter_seq = filter(result, seq)

for i in filter_seq:
    print(i)
```

```python
seq = [0, 1, 2, 3, 4, 5, 6]
result = lambda x: x % 2 != 0

filter_seq = filter(result, seq)

print(list(filter_seq))
```

# Exceptions in python

## What is exception?
- Exception is an event that occurs during an execution/running a code/program this basically disrupt the flow of the execution/running of code. 

## Standard exceptions in  python:

1. FileNotFound
2. ImportError
3. RuntimeError
4. NameError
5. TypeError

## Why use exception?
- Not to allow errors and run remaining part of the program.
- Helps in debugging(handle errors in the program and correct them)
- Gives a cleaner code
## What are errors and types?
- Action that cause/gives incoorect result.
Ex: syntax error
> How to handle these errors?

#### try and except block 

```python
a = 10
b = 0
c = a/b 
print(c)
# We have this error
```

```python
try:
    a = 10
    b = 0
    c = a/b
    print(c)
except:
    print("Cannot divide any int number with a zero, Please enter positive value")
```

```python
try:
    a  = 10
    b  = 0

    c = a/b
    print(c)
except ValueError:
    print('Please enter integer number, not a string')
except ZeroDivisionError:
    print('Cannot divide any int a numer with zero, Please enter a positive integer')
```

```python
try:
    a  = 10
    b  = 0
    c = a/b
    print(c)
except (ValueError,ZeroDivisionError):
    print('Please enter integer number, not a string, and do please enter a positive integer')
```

```python
try:
    a  = int(input())
    b  = int(input())
    c = a/b
    print(c)
except (ValueError,ZeroDivisionError):
    print('Please enter integer number, not a string, and do please enter a positive integer')
else:
    print('Code Ran Successfully')
```

## Copy & Views
- The main difference between a copy and a view of an array is that the copy is a new array, and the view is just a view of original array.
- The copy owns the data and any changes made to the copy will not affect original array, and any changes made to the original array will not affect the copy.
- The view does not own the data and any changes made to the view will affect the original array, and any changes made to the original array will affect the view. 

```python
## Shallow copy

arr2 = arr1.copy()
arr2
```
